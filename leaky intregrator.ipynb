```python
#leaky intregretor
import numpy as n
import random as r
import matplotlib.pyplot as plt
sig = [5*n.sin(2*n.pi*t) for t in n.linspace(0,1,1000)]
noise = [r.random()-0.5 for i in n.linspace(0,1,1000)]
xn = [i+j for i,j in zip(sig,noise)]
m=100
l=m-1/m
xnn=list(n.zeros(m-1))+list(xn)

yn = [l*(sum(xnn[k-m+1:k])/len(xnn[k-m+1:k]))+(1-l)*xnn[k] for k in n.arange(m-1,len(xnn))]

plt.plot(xn)
plt.plot(yn)
plt.show()
```

```python
import numpy as n
import random as r
import matplotlib.pyplot as plt
sig = [5*n.sin(2*n.pi*t) for t in n.linspace(0,1,1000)]
noise = [r.random()-0.5 for i in n.linspace(0,1,1000)]
xn = [i+j for i,j in zip(sig,noise)]
yn = [0]
l=m-1/m
xnn = list(n.zeros(m-1)+list(xn))
for i in n.arange(m-1,len(xn)):
    yn.append(l*yn(i-1-(m-2)+(1-l)+xnn[i]))
    
```

```python
import numpy as np
x = np.array([1,8,0,4,3,2,2])
len(x)
```

```python
N = (len(x)-1)//2 
n = np.array([i for i in range(-N,N+1)])     # n = -N,.....N
N,n
```

```python
import matplotlib.pyplot as plt
L = []   # all the interploter (Lm3,Lm2.........L3)
# term Ln = prod((t-k)/(n-k));  k=-N.....N    k!=n
l=[]
tim=[i for i in np.linspace(-N,N,1000)]    #input in t
n = range(-N,N+1)
for i in n:    # it define individual interplotors
    for t in tim:
        term=1
        for k in range(-N,N+1):  
            if k!=i:
                term = term*((t-k)/(i-k))
        l.append(term)
    L.append(l)
    l=[]
for i in range(len(L)):
    plt.plot(tim,L[i])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
L = []   # all the interploter (Lm3,Lm2.........L3)
x = [1,8,0,4,3,2,2]
l = []
tim = [i for i in np.linspace(-3,3,1000)]    #input in t
n = [-3,-2,-1,0,1,2,3]
for i in n:    # it define individual interplotors
    for t in tim:
        term = 1
        for k in range(-N, N+1): 
            if k!=i:
                term = term*((t-k)/(i-k))   # k = -2,-1,0,1,2,3
        l.append(term)
    L.append(l)
    l=[]
newL = [i*np.array(j) for i,(j) in zip(x,L)]
y_ = sum(newL)
plt.plot(tim, y_)
plt.show()
```

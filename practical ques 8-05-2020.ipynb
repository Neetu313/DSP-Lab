```python
# ques 1...Create a mixture of 3 sinusoids of different linear frequencies which will be you date of birth.Then find Fourier
#transform to identify the frequencies.
import thinkdsp as t
import thinkplot as d

a = t.SinSignal(freq = 12, amp = 5, offset = 0)
a_wave=a.make_wave(duration = 1.5, start = 0, framerate = 42000)
a_spec = a_wave.make_spectrum()

b = t.SinSignal(freq = 7, amp = 5, offset = 0)
b_wave = b.make_wave(duration = 1.5, start = 1.5, framerate = 42000)
b_spec = b_wave.make_spectrum()

c = t.SinSignal(freq = 2000, amp = 5, offset = 0)
c_wave = c.make_wave(duration = 1.5,start = 3,framerate = 42000)
c_spec = c_wave.make_spectrum()

ABC = a_wave + b_wave + c_wave
aud = ABC.make_audio()
aud
```

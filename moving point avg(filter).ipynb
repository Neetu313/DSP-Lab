```python
#create signal using numpy array
import matplotlib.pyplot as mp
import numpy as np
from random import random
sig=[5*np.sin(2*np.pi*t) for t in np.linspace(0,1,1000)] 
noise=[-0.5+random()*1 for i in np.arange(len(sig))]
xn=sig+noise
len(np.linspace(0,1,1000))
mp.plot(xn)
mp.show()
```

```python
#make noise signal
import numpy as np
x=[i+j for i , j in zip(sig,noise)]

m=250
xnn=list(np.zeros(m-1))+x
yn=[sum(xnn[n-m+1:n+1])/m for n in np.arange(m-1,len(xn))]
mp.plot(x)
mp.plot(yn)
mp.show()
```

```python
#moving point avg
yn=[]
for i in np.arange(1000):
    term=x[i]+x[i-1]+x[i-2]/3
    yn.append(term)
mp.plot(yn)        
```
